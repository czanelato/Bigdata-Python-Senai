    O grafico de linha gerado deve incluir as seguintes propriedades de estilo:

    O estilo da linha pontilhada e a cor da linha devem ser vermelhas

    Mostrar legenda no local inferior direito.

    Nome do rotulo X = Numero do mes

    Nome do rotulo Y = numero de unidades vendidas

    Adicione um marcador de circulo.

    Cor do marcador de linha conforme lida

    A largura da linha deve ser 3



```python
# Importar as bibliotecas 
import pandas as pd
import matplotlib.pyplot as plt
import os

# Mapear o caminho do arquivo
caminhoArquivo = "C:\\Users\\FIC\\Documents\\Carlos Zanelato_Python\\Arquivos\\company_sales_data.csv"

# Alocar os dados em um Dataframe
df = pd.read_csv(caminhoArquivo)

# Visualizar o cabealho do Dataframe
df.head()
```

```python
# Selecionar os dados que vo ser apresentados no grfico

totalVendido = df['total_units'].to_list()
listameses = df['month_number'].to_list()

# Criar o grfico utilizando os dados listados anteriormente
plt.plot(
    listameses, # Eixo X
    totalVendido, # Eixo Y
    label = 'Total Vendido por Ms', # Conteudo da legenda
    c = 'red', # Cor da linha
    marker = 'o', #Marcador
    markerfacecolor = 'k', # Cor interna do marcdor
    ls = '--', # Tipo de linha
    linewidth = 3 # Espessura da linha
    )

plt.xlabel('Meses') # Ttulo do eixo X
plt.ylabel('Vendas em dlares') # Ttulo do eixo Y
plt.title('Vendas por Ms do ano anterior') # Ttulo do grfico
plt.yticks([10000,20000,30000,40000,50000]) # Ajustando os intervalos do eixo Y
plt.yticks(listameses) # Ajustando os intervalos do eixo X

plt.legend(loc = 'lower right')
```
