# Primeiros Passos com Matplotlib

```python
# Importanto as blibliotecas

import matplotlib.pyplot as plt
import numpy as np

# Criando os arrays (matrizes) para utilizar no gráficos

x = np.array([1,2,3,4,5])
y = np.array([2,4,6,8,10])

# Criando um arquivo com os valores anteriores
plt.plot(x, y)
```

# MODIFICANDO O TIPO DE MARCADOR

```python
plt.plot(x,y, marker = 'X', c = 'red')
```

# Modificando a forma como os pontos são ligados e o tipo de de linha

```python
plt. plot(x,
          y,
          marker = '*',
          c = 'blue',
          ds = 'steps', # forma de ligação
          ls = '--') # estilo de linha)
```

# Alterando as cores dos marcadores

```python
plt. plot(x, # Vetor responsável pelos valores do eixo X
          y, # Vetor responsável pelo valores do eixo Y
          marker = '*', # Marcador do gráfico
          c = 'red', # Cor da linha
          ds = 'steps', # forma de ligação
          ls = '--' ,# estilo de linha
          mec = 'blue' # Cor dos marcadores
)
```

# Adicionando outros elementos ao gráfico

```python
plt.plot(x,y) # Criando o gráfico

plt.xlabel('Valores de X', fontsize = 12) # Adicionando o título do eixo X

plt.ylabel('Valores de Y', fontsize = 12) # Adicionando o título do eixo Y

plt.title('Título do Gráfico', fontsize = 20, c = 'red') # Adicionando o título do gráfico

plt.xticks(fontsize = 8, c = 'blue') # Alterando as cores dos rótulos de X e Y
plt.yticks(fontsize = 8, c = 'purple')
```

# Exemplo 1 - Gráfico de Pizza

```python
# Importar as Bibliotecas
import matplotlib.pyplot as plt

# Dados para plotar no gráfico
languages= ['java', 'python', 'php', 'javascript', 'c#', 'c++']
popularidade= [22.2,17.6,8.8,8,7.7,6.7]

# Explosão das Fatias
explosao = [0.1,0,0,0,0,0]

# Definindo cores
cores = ['#16E1F0','#1657F0','#169CF0','#11F5B8','#7775F0','#F02C17']

# Gerar o gráfico
# labels = rótulos de categoria | autopct = rótulos numéricos
# %0.1%% > formatando os valores dos rótulos numéricos com 1 casa após a virgula
# shadow > Sombra
plt.pie(popularidade, labels= languages, autopct='%1.0f%%', shadow= True,
        explode = explosao,
        colors= cores )
```

```python
# Criando os arrays (matrizes) para utilizar no gráficos

x = np.array([-3,-2.5,-2,-1,0,1,2,2.5,3])
y = np.array([-27,-15.62,-8,-1,0,1,8,15.62,27])

# Criando um arquivo com os valores anteriores
plt.plot(x, y, label = 'Função Cúbica', marker = 'o', color = 'magenta', ls= '--', lw= 3)

# Adicionar um título aos eixos
plt.xlabel('Valor de X', fontsize = 15)
plt.ylabel('Valor de Y', fontsize = 15.5)

plt.title('TÍTULO DO GRÁFICO')



```

# Seaborn

Seaborn é uma biblioteca baseada em Matplolib que traz uma interface mais atraente para os gráficos, além de facilitar a modificação de parâmetros default do Matplotlib e permite a utilização de dataframes.

```python
# Importando as bibliotecas
import seaborn as sns
import matplotlib.pyplot as plt
from pylab import randn

# Geração de 200 dados aleatórios
x = randn(200)
y = randn(200)

# Criar o gráfico de dispersão
sns.scatterplot(x = x,y = y)

```

```python
# Importando as bibliotecas
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Mapear o caminho do arquivo
caminhoArquivo = "C:\\Users\\FIC\\Documents\\Carlos Zanelato_Python\\Arquivos\\sea-level.csv"

# Alocar os dados em um Dataframe
dados = pd.read_csv(caminhoArquivo)

# Visualizar o cabeçalho
# (Selecionar a partir da linha 5(5:) e todas as colunas(:))
dados = dados.loc[6:,:]

dados.columns = ['Year',
                 'CSIRO - Adjusted sea level (inches)',
                 'CSIRO - Lower error bound (inches)',
                 'CSIRO - Upper error bound (inches)',
                 'NOAA - Adjusted sea level (inches)'
                 ]

dados.head()

# Converter os dados em numéricos
dados['Year'] = dados['Year'].astype(int)
dados['CSIRO - Adjusted sea level (inches)'] = dados['CSIRO - Adjusted sea level (inches)'].astype(float)
dados['CSIRO - Lower error bound (inches)'] = dados['CSIRO - Lower error bound (inches)'].astype(float)
dados['CSIRO - Upper error bound (inches)'] = dados['CSIRO - Upper error bound (inches)'].astype(float)
dados['NOAA - Adjusted sea level (inches)'] = dados['NOAA - Adjusted sea level (inches)'].astype(float)

dados.info()
```

```python
anos = dados['Year'].to_list()
nivel = dados['CSIRO - Adjusted sea level (inches)'].to_list()

sns.scatterplot(x = anos, y = nivel)
```
