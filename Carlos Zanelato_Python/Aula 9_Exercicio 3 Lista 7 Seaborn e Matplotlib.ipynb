    Leia todos os dados de vendas de produtos e mostre-os usando um grfico multilinha.
    Exiba o nmero de unidades vendidas por ms para cada produto usando grficos multilinhas. (ou seja, faa a plotagem das linhas separadas para cada produto).



```python
# Importar as bibliotecas 
import pandas as pd
import matplotlib.pyplot as plt
import os

# Mapear o caminho do arquivo
caminhoArquivo = "C:\\Users\\FIC\\Documents\\Carlos Zanelato_Python\\Arquivos\\company_sales_data.csv"

# Alocar os dados em um Dataframe
df = pd.read_csv(caminhoArquivo)

# Visualizar o cabealho do Dataframe
df.head()
```

```python
# Criar listas com base nas colunas do Dataframe

listameses = df['month_number'].to_list()
faceCreamSalesData = df['facecream'].to_list()
faceWashSalesData = df['facewash'].to_list()
toothPasteSalesData = df['toothpaste'].to_list()
bathingSoapSalesData = df['bathingsoap'].to_list()
shampooSalesData = df['shampoo'].to_list()
moisturizerSalesData = df['moisturizer'].to_list()

plt.plot(listameses, faceCreamSalesData, label = 'Face Cream', c = 'blue', marker = 'o', linewidth = 3)
plt.plot(listameses, faceWashSalesData, label = 'Face Wash', c = 'red', marker = 'o', linewidth = 3)
plt.plot(listameses, toothPasteSalesData, label = 'Tooth Paste', c = 'yellow', marker = 'o', linewidth = 3)
plt.plot(listameses, bathingSoapSalesData, label = 'Bathing Soap', c = 'green', marker = 'o', linewidth = 3)
plt.plot(listameses, shampooSalesData, label = 'Shampoo', c = 'purple', marker = 'o', linewidth = 3)
plt.plot(listameses, moisturizerSalesData, label = 'Moisturizer', c = 'blue', marker = 'o', linewidth = 3)

plt.xlabel('Month Numer')
plt.ylabel('Sales Units in Number')

plt.legend(loc = 'upper left')

plt.xticks(listameses)
plt.yticks([1000,2000,4000,8000,10000,12000])
```
