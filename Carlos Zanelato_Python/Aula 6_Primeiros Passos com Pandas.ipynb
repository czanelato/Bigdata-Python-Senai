# Renomear as colunas

```python
# Renomear todas as colunas
# As colunas devem estar na mesma ordem dos nomes
# A quantidade deve ser igual a quantidade de colunas

df.columns=['idade','job','estado civil','educação','defaut','balance','housing','emprestimo','contato','dia','mes','duração','campanha','pdays','previous','poutcome']

df.head()
```
